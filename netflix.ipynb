````python
#Import all necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
````

<i>Creating a list of years, dating from 2011 - 2020 and an average random movie length.
Thereafter, a the years and movie lengths are compiled into a dictionary named <code>'movie_dict'</code> and then converted into a dataframe, <code>'durations_pf'</code> and then displayed.</i>

````python
#Creating a range of values from 2011 - 2020
list_years = list(range(2011, 2021))

#Random movie length
length = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]

print(f'Year List: \n {list_years}')
print(f'Length: \n {length}')
````

````python
#A dictionary to house the year list and movie lenghts
movie_dict = {"years": list_years, "durations": length}
print(movie_dict)
````

````python
#Converting the dictionary into a dataframe
durations_df = pd.DataFrame(movie_dict)
display(durations_df)
````

## Visualisations of the duration_df dataframe

Line plot performed using seaborn library with <code>'Years'</code> on the x-axis and <code>'Durations'</code> on the y-axis, titled <code>"Netflix Movie Durations 2011 - 2020"</code>.

````python
fig = plt.subplots(figsize=(15, 10))
sns.lineplot(
    x = durations_df['years'], y = durations_df['durations'], data = durations_df, marker = '^', markerfacecolor = 'indigo',
    color = 'cyan').set(title = "Netflix Movie Durations 2011 - 2020", xlabel = 'Years', ylabel = 'Duration (min)')
sns.set_theme(style = 'darkgrid', font_scale = 1)
````

**Filter by Movies**

```python
After a preview of the data, it was observed that there are two types of netflix shows, namely: "TV Shows" and "Movies".
Having to filter the dataset using the .loc[] function to gain insight on the Movie type. A new dataframe is created which contains: title, country, release year, duration, genre.

Let's have at it!!
```

````python
#Subset the dataframe for movies
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']

#Pick the features
netflix_movies_col_subset = ['title', 'country', 'genre', 'release_year', 'duration']
netflix_movies_col_subset = netflix_df_movies_only[netflix_movies_col_subset]


#Print the first 5 rows of the column subset dataframe
display(netflix_movies_col_subset.head())
````

### **Create a Scatter Plot**

```python
Visualise the realease_year on the x-axis vs movie duration on the y-axis, titled "Movie Duration by Year of Realease" with a scatter plot using seaborn visualisation module.
```

````python
fig = plt.subplots(figsize=(20, 10))
sns.scatterplot(
    data = netflix_movies_col_subset, x = netflix_df_movies_only['release_year'], 
    y = netflix_df_movies_only['duration'],hue = None, style = None).set(
        title = "Movie Duration by Year of Release", xlabel = 'Realease Year', ylabel = 'Duration (min)')
````

````python
fig = px.scatter(netflix_movies_col_subset, x="release_year", y="duration")
#Edit the layout
fig.update_layout(title = "Movie Duration by Year of Release", 
                  xaxis_title='Release Year', yaxis_title='Duration (min)')
fig.show()
````

### Marking non-feature films

<p>Append Colors to genre by using a <code>for</code> loop to <i>iterate</i> through the rows and conditional statments to <code>append</code> the colors per genre based on <code>'Children'</code>, <code>'Documentaries'</code>, <code>'Stand-Up'</code>, <code>'Others'</code>.</p>

````python
#Create an empty colors list
colors = []

#Use a for loop to iterate the netflix_movie_col_subset rows
for index, row in netflix_movies_col_subset.iterrows():
    genre = row['genre']
    if genre == 'Children':
        colors.append('red')
    elif genre == 'Documentaries':
        colors.append('blue')
    elif genre == 'Stand-Up':
        colors.append('green')
    else:
        colors.append('black')

#Inspect the result by printing the first 10 values
colors[0:10]
````

<p>Plot a scatter plot with labels <code>"Release Years"</code> on the x-axis and <code>"Duration (min)"</code> on the y-axis using the <code>colors</code> introduced earlier for the genres</p>

````python
fig = plt.subplots(figsize=(15, 10))
sns.set_palette(colors)
sns.scatterplot(
    data = netflix_movies_col_subset, x = netflix_df_movies_only['release_year'], 
    y = netflix_df_movies_only['duration'], color = colors).set(
    title = "Movie Duration by Year of Release", xlabel = 'Realease Year', ylabel = 'Duration (min)')
````
